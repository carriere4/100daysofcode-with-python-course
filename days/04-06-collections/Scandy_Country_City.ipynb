```python
from collections import defaultdict, namedtuple, Counter
import csv
```

```python
cities_csv = 'country_city_model.csv'
```

```python
Country = namedtuple('Country', ('date city model'))
```

```python
def get_users_by_country(data):
    countries = defaultdict(list) #create default directory of countries
    with open(data) as f:
        for line in csv.DictReader(f):
            try:
                date = line['Date']
                country = line['Country']
                city = line['City']
                model = line['Model']
            except:
                ValueError
                continue
            c = Country(date=date, city=city, model=model)
            countries[country].append(c)
    return countries       
```

```python
my_countries = get_users_by_country(cities_csv)
```

```python
my_countries['United States']
```

```python
cnt = Counter()
```

```python
for country, city in my_countries.items():
    cnt[country] += len(city)
```
